```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("sinannet_cnv_animaisbr161447179_97_224_142.csv", sep=";",encoding="latin1")
df1 = pd.read_csv("acid_animais_pec_2024.csv", sep=";",encoding="latin1")
df2 = pd.read_csv("acid_animais_pec_2023.csv", sep=";",encoding="latin1")
df3 = pd.read_csv("acid_animais_pec_2022.csv", sep=";",encoding="latin1")
```

```python
desc_2025 = df["Total"].describe()
desc_2024 = df1["Total"].describe()
desc_2023 = df2["Total"].describe()
desc_2022 = df3["Total"].describe()

tabela_describe = pd.concat([desc_2025, desc_2024, desc_2023, desc_2022], axis=1)

tabela_describe.columns = ["2025", "2024", "2023", "2022"]

tabela_describe

```

```python
meses = ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun", "Jul", "Ago", "Set", "Out", "Nov", "Dez"]

# Para o ano de 2025

for mes in meses:
    df[mes] = pd.to_numeric(df[mes].replace('-', 0), errors ="coerce")
    
tabela_mensal = df[meses].sum()
tabela_mensal = tabela_mensal.reset_index()
tabela_mensal.columns = ["Mês", "Total"]
print(tabela_mensal)

plt.figure()
plt.plot(tabela_mensal["Mês"], tabela_mensal["Total"], marker="o")
plt.xlabel("Mês")
plt.ylabel("Total de registros")
plt.title("Total mensal de acidentes por animais peçonhentos - 2025")
plt.show()
```

```python
# Para o ano de 2024
for mes in meses:
    df1[mes] = pd.to_numeric(df1[mes].replace('-', 0), errors ="coerce")
    
tabela_mensal_1 = df1[meses].sum()
tabela_mensal_1 = tabela_mensal_1.reset_index()
tabela_mensal_1.columns = ["Mês", "Total"]
print(tabela_mensal_1)

plt.figure()
plt.plot(tabela_mensal_1["Mês"], tabela_mensal_1["Total"], marker="o")
plt.xlabel("Mês")
plt.ylabel("Total de registros")
plt.title("Total mensal de acidentes por animais peçonhentos - 2025")
plt.show()
```

```python
# Para o ano de 2023

for mes in meses:
    df2[mes] = pd.to_numeric(df2[mes].replace('-', 0), errors ="coerce")
    
tabela_mensal_2 = df2[meses].sum()
tabela_mensal_2= tabela_mensal_2.reset_index()
tabela_mensal_2.columns = ["Mês", "Total"]
print(tabela_mensal_2)

plt.figure()
plt.plot(tabela_mensal_2["Mês"], tabela_mensal_2["Total"], marker="o")
plt.xlabel("Mês")
plt.ylabel("Total de registros")
plt.title("Total mensal de acidentes por animais peçonhentos - 2025")
plt.show()
```

```python
# Para o ano de 2022

for mes in meses:
    df3[mes] = pd.to_numeric(df3[mes].replace('-', 0), errors ="coerce")
    
tabela_mensal_3 = df3[meses].sum()
tabela_mensal_3= tabela_mensal_3.reset_index()
tabela_mensal_3.columns = ["Mês", "Total"]
print(tabela_mensal_3)

plt.figure()
plt.plot(tabela_mensal_3["Mês"], tabela_mensal_3["Total"], marker="o")
plt.xlabel("Mês")
plt.ylabel("Total de registros")
plt.title("Total mensal de acidentes por animais peçonhentos - 2025")
plt.show()
```

```python
fig = plt.figure(figsize=(12,7))
df = fig.add_subplot(2, 2, 1)
df1 = fig.add_subplot(2, 2, 2)
df2= fig.add_subplot(2, 2, 3)
df3 = fig.add_subplot(2, 2, 4)

df.plot(tabela_mensal["Mês"], tabela_mensal["Total"], marker="o", color='blue', label='2025')
df.set_title("2025")

df1.plot(tabela_mensal_1["Mês"], tabela_mensal_1["Total"], marker="o", color='green', label='2024')
df1.set_title("2024")

df2.plot(tabela_mensal_2["Mês"], tabela_mensal_2["Total"], marker="o", color='red', label='2023')
df2.set_title("2023")

df3.plot(tabela_mensal_3["Mês"], tabela_mensal_3["Total"], marker="o", color='orange', label='2022', )
df3.set_title("2022")

fig.suptitle("Distribuição mensal de acidentes por animais peçonhentos(2022 - 2025)", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.98])

plt.show()
```
